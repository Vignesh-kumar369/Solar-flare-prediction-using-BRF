```python
import pandas as pd
import numpy as np

# --- Configuration ---
INPUT_FILE = 'model_ready_dataset.parquet'
OUTPUT_FILE = 'model_ready_feature_engineered.parquet'

print(f"--- Loading dataset: {INPUT_FILE} ---")
df = pd.read_parquet(INPUT_FILE)
df['T_REC'] = pd.to_datetime(df['T_REC'])
df.sort_values(by=['NOAA_AR', 'T_REC'], inplace=True)

# --- Feature Engineering ---
print("\n--- Engineering new time-series features ---")

# Set the timestamp as the index, which is needed for time-series operations
df.set_index('T_REC', inplace=True)

# List of features to engineer
features_to_engineer = [
    'USFLUX', 'MEANGAM', 'MEANGBT', 'MEANGBZ', 'MEANGBH', 'TOTUSJH',
    'TOTPOT', 'MEANPOT', 'TOTUSJZ', 'SAVNCPP', 'ABSNJZH', 'AREA_ACR',
    'MEANJZH', 'R_VALUE', 'MEANALP', 'MEANSHR'
]

# We will calculate the rate of change (derivative) over 6 and 12 hours
time_windows = ['6h', '12h']

# Group by each Active Region to calculate features independently
grouped = df.groupby('NOAA_AR')

for feature in features_to_engineer:
    for window in time_windows:
        # Calculate the difference over the time window
        df[f'{feature}_d{window}'] = grouped[feature].diff(periods=int(pd.Timedelta(window).total_seconds() / 720))

# After creating derivatives, we have NaNs at the start of each AR's history.
# We should drop these rows as they don't have complete feature information.
df.dropna(inplace=True)

# Reset the index to turn T_REC back into a column
df.reset_index(inplace=True)

print(f"New feature-engineered dataset has {len(df.columns)} columns.")
print(f"Saving new dataset to {OUTPUT_FILE}...")
df.to_parquet(OUTPUT_FILE, index=False)

print("\n✅ Feature engineering complete!")
```

```python
import pandas as pd
import numpy as np
from numba import jit

# =========================
# CONFIGURATION
# =========================
# Ensure this input file contains the 6h/12h derivatives and CORRECTED labels
INPUT_FILE = "model_ready_feature_engineered.parquet"
OUTPUT_FILE = "model_ready_with_stats_OPTIMIZED.parquet"

# Base SHARP features (The original 16 features)
base_features = [
    'USFLUX', 'MEANGAM', 'MEANGBT', 'MEANGBZ', 'MEANGBH', 'TOTUSJH',
    'TOTPOT', 'MEANPOT', 'TOTUSJZ', 'SAVNCPP', 'ABSNJZH', 'AREA_ACR',
    'MEANJZH', 'R_VALUE', 'MEANALP', 'MEANSHR'
]

# Derivative features already in file
lagged_features = [f'{col}_{suffix}' for col in base_features for suffix in ['d6h', 'd12h']]

ROLLING_WINDOW = 3 # The size of the rolling window (3 time steps = 36 minutes)
ADD_SLOPE = True 

# =========================
# 1. NUMBA SLOPE FUNCTION (The Optimization)
# =========================
@jit(nopython=True)
def slope_func_numba(y):
    """
    Numba-accelerated function to calculate the slope (linear trend) of a 
    short series (e.g., 3 periods), significantly faster than np.polyfit inside apply.
    """
    if len(y) < 2:
        return np.nan
    
    t = np.arange(len(y))
    n = len(y)
    
    # Calculate components for linear regression (y = m*t + c)
    sum_t = np.sum(t)
    sum_y = np.sum(y)
    sum_ty = np.sum(t * y)
    sum_tt = np.sum(t * t)
    
    # Calculate slope (m) using the analytic solution
    denominator = (n * sum_tt) - (sum_t * sum_t)
    if denominator == 0:
        return np.nan
    
    return (n * sum_ty - sum_t * sum_y) / denominator

# =========================
# 2. LOAD AND PREPARE DATA
# =========================
print(f"--- Loading {INPUT_FILE} ---")
df = pd.read_parquet(INPUT_FILE).copy()
df['T_REC'] = pd.to_datetime(df['T_REC'])
print(f"Loaded {len(df):,} rows, {len(df.columns)} columns")

# Sort once for correct rolling
df = df.sort_values(["HARPNUM", "T_REC"]).reset_index(drop=True)

# Group by HARPNUM once for all calculations
grouped = df.groupby("HARPNUM")

# =========================
# 3. GENERATE ALL NEW FEATURES EFFICIENTLY
# =========================
print(f"\n--- Generating rolling features (window={ROLLING_WINDOW}) ---")
new_features_list = []
slope_features_list = []

for col in lagged_features:
    if col in df.columns:
        # Standard Rolling Stats (Pandas is fast for these)
        roll = grouped[col].rolling(ROLLING_WINDOW, min_periods=1)
        stats = pd.DataFrame({
            f"{col}_mean{ROLLING_WINDOW}": roll.mean().reset_index(level=0, drop=True),
            f"{col}_std{ROLLING_WINDOW}": roll.std().reset_index(level=0, drop=True),
            f"{col}_min{ROLLING_WINDOW}": roll.min().reset_index(level=0, drop=True),
            f"{col}_max{ROLLING_WINDOW}": roll.max().reset_index(level=0, drop=True)
        })
        new_features_list.append(stats)
        
        # Numba-Accelerated Slope
        if ADD_SLOPE:
            slope = grouped[col].rolling(ROLLING_WINDOW, min_periods=2).apply(slope_func_numba, raw=True)
            slope_features_list.append(pd.DataFrame({
                f"{col}_slope{ROLLING_WINDOW}": slope.reset_index(level=0, drop=True)
            }))
    else:
        print(f"⚠ Skipping {col}, not in dataset")

# =========================
# 4. COMBINE, CLEAN, AND SAVE
# =========================
print("\n--- Combining original data with new features ---")

# Combine the original DataFrame with all standard stats
df = pd.concat([df] + new_features_list, axis=1)

# Combine with all slope stats
if slope_features_list:
     df = pd.concat([df] + slope_features_list, axis=1)

# Fill NAs created by the rolling windows with 0 (safer than dropping for imbalanced data)
print("\n--- Finalizing dataset (filling NaNs with 0) ---")
df = df.fillna(0)
print(f"Final shape: {df.shape}")

df.to_parquet(OUTPUT_FILE, index=False)
print(f"\n✅ Successfully generated and saved OPTIMIZED dataset to {OUTPUT_FILE}")
```

```python
# feature_engineering_v3_advanced.py
import pandas as pd
import numpy as np

# =========================
# CONFIG
# =========================
INPUT_FILE = "model_ready_feature_engineered.parquet"
OUTPUT_FILE = "features_v3_advanced.parquet"
GROUP_COL = "HARPNUM"
TIME_COL = "T_REC"

# --- Define the features to create ---
FEATURES_TO_ENHANCE = [
    'TOTPOT',
    'TOTUSJH',
    'TOTUSJZ',
    'USFLUX',
    'MEANSHR'
]
# Define time windows in hours. 24 hours is a good start.
WINDOWS_IN_HOURS = [24]
# ------------------------------------

# =========================
# SCRIPT
# =========================
print(f"➡️ 1. Loading data from '{INPUT_FILE}'...")
df = pd.read_parquet(INPUT_FILE)
df[TIME_COL] = pd.to_datetime(df[TIME_COL])
df = df.sort_values(by=[GROUP_COL, TIME_COL])
print(f"✅ Data loaded with {len(df)} rows.")

# Calculate how many 12-minute periods are in each hour window
periods_map = {f"{hr}hr": hr * 5 for hr in WINDOWS_IN_HOURS}

# =========================
# 2. CREATE ADVANCED FEATURES
# =========================
print("\n➡️ 2. Engineering advanced physics-informed features...")
original_feature_count = len(df.columns)

for feat in FEATURES_TO_ENHANCE:
    for hr_str, periods in periods_map.items():
        # --- Create Rate of Change (Delta) ---
        delta_col = f'{feat}_delta_{hr_str}'
        df[delta_col] = df.groupby(GROUP_COL)[feat].diff(periods=periods)
        
        # --- Create "High-State Decay" Interaction Feature ---
        interaction_col = f'High_State_Decay_{feat}_{hr_str}'
        df[interaction_col] = df[feat] * df[delta_col]

# --- Create Physics-Informed Ratios ---
print("   - Calculating ratio features...")
df['HELICITY_per_FLUX'] = df['TOTUSJH'] / (df['USFLUX'] + 1e-9)
df['VERT_CURRENT_per_FLUX'] = df['TOTUSJZ'] / (df['USFLUX'] + 1e-9)


new_features_created = len(df.columns) - original_feature_count
print(f"✅ Created {new_features_created} new features.")

# =========================
# 3. CLEANUP AND SAVE
# =========================
print("\n➡️ 3. Cleaning and saving new dataset...")
# The new features will create NaN values where the window is not yet full. We drop these.
rows_before_drop = len(df)
df.dropna(inplace=True)
rows_after_drop = len(df)
print(f"   - Dropped {rows_before_drop - rows_after_drop} rows with NaN values from window calculations.")

# Reset the index before saving
df.reset_index(drop=True, inplace=True)

df.to_parquet(OUTPUT_FILE)
print(f"\n✅ Successfully saved enhanced dataset to '{OUTPUT_FILE}'.")
```

```python
# --- STEP 1: create_deltas_ratios.py ---
import pandas as pd
import numpy as np

# --- CONFIG ---
INPUT_FILE = "final_labeled_2014-2015_dataset.parquet"
OUTPUT_FILE = "intermediate_deltas_ratios.parquet"
GROUP_COL = "HARPNUM"
TIME_COL = "T_REC"
TIME_WINDOWS = ['6h', '12h']

BASE_SHARP_FEATURES = [
    'USFLUX', 'MEANGAM', 'MEANGBT', 'MEANGBZ', 'MEANGBH', 'TOTUSJH',
    'TOTPOT', 'MEANPOT', 'TOTUSJZ', 'SAVNCPP', 'ABSNJZH', 'AREA_ACR',
    'MEANJZH', 'R_VALUE', 'MEANALP', 'MEANSHR'
]

print(f"--- Running Step 1: Creating Deltas and Ratios ---")
df = pd.read_parquet(INPUT_FILE).copy()
df[TIME_COL] = pd.to_datetime(df[TIME_COL])
df = df.sort_values(by=[GROUP_COL, TIME_COL]).reset_index(drop=True)

# --- A. Physics-Informed Ratios ---
df['HELICITY_per_FLUX'] = df['TOTUSJH'] / (df['USFLUX'] + 1e-9)
df['VERT_CURRENT_per_FLUX'] = df['TOTUSJZ'] / (df['USFLUX'] + 1e-9)
new_base_features = BASE_SHARP_FEATURES + ['HELICITY_per_FLUX', 'VERT_CURRENT_per_FLUX']

# --- B. Temporal Deltas ---
grouped = df.groupby(GROUP_COL)
for feature in new_base_features:
    for window in TIME_WINDOWS:
        periods = int(pd.Timedelta(window).total_seconds() / 720)
        delta_col = f'{feature}_d{window}'
        df[delta_col] = grouped[feature].diff(periods=periods)

print(f"Saving intermediate result to {OUTPUT_FILE}")
df.to_parquet(OUTPUT_FILE, index=False)
print("✅ Step 1 Complete. Memory flushed.")
```

```python
# --- STEP 2: create_rolling_stats.py ---
import pandas as pd
import numpy as np
from numba import jit

# --- CONFIG ---
INPUT_FILE = "intermediate_deltas_ratios.parquet"
OUTPUT_FILE = "intermediate_all_stats.parquet"
GROUP_COL = "HARPNUM"
ROLLING_WINDOW = 3

# --- NUMBA SLOPE FUNCTION (Must be defined here for speed) ---
@jit(nopython=True)
def slope_func_numba(y):
    if len(y) < 2: return np.nan
    t = np.arange(len(y))
    n = len(y)
    sum_t = np.sum(t)
    sum_y = np.sum(y)
    sum_ty = np.sum(t * y)
    sum_tt = np.sum(t * t)
    denominator = (n * sum_tt) - (sum_t * sum_t)
    if denominator == 0: return np.nan
    return (n * sum_ty - sum_t * sum_y) / denominator
# -------------------------------------------------------------

print(f"--- Running Step 2: Generating Rolling Stats ---")
df = pd.read_parquet(INPUT_FILE).copy()
grouped = df.groupby(GROUP_COL)

# --- Find all delta/ratio features created in Step 1 for rolling ---
lagged_features = [c for c in df.columns if c.endswith(('_d6h', '_d12h'))]

new_features_list = []
for col in lagged_features:
    if col in df.columns:
        roll = grouped[col].rolling(ROLLING_WINDOW, min_periods=1)
        
        stats = pd.DataFrame({
            f"{col}_mean{ROLLING_WINDOW}": roll.mean().reset_index(level=0, drop=True),
            f"{col}_std{ROLLING_WINDOW}": roll.std().reset_index(level=0, drop=True),
            f"{col}_min{ROLLING_WINDOW}": roll.min().reset_index(level=0, drop=True),
            f"{col}_max{ROLLING_WINDOW}": roll.max().reset_index(level=0, drop=True)
        })
        new_features_list.append(stats)
        
        slope = grouped[col].rolling(ROLLING_WINDOW, min_periods=2).apply(slope_func_numba, raw=True)
        new_features_list.append(pd.DataFrame({
            f"{col}_slope{ROLLING_WINDOW}": slope.reset_index(level=0, drop=True)
        }))

df = pd.concat([df] + new_features_list, axis=1)

print(f"Saving intermediate result to {OUTPUT_FILE}")
df.to_parquet(OUTPUT_FILE, index=False)
print("✅ Step 2 Complete. Memory flushed.")
```

```python
import pyarrow.parquet as pq
import pyarrow as pa
import pandas as pd
import numpy as np
import os
import gc

# =========================
# CONFIGURATION
# =========================
INPUT_FILE = "intermediate_all_stats.parquet"
OUTPUT_FILE = "final_arsenal_300_features_OPTIMIZED.parquet"
BATCH_SIZE = 10000  # Strict limit: Only 10k rows at a time (Tiny RAM footprint)

print(f"--- Running Step 3: Ultra-Safe Batch Processing ---")

if not os.path.exists(INPUT_FILE):
    raise FileNotFoundError(f"Input file {INPUT_FILE} not found. Did Step 2 finish?")

# 1. Open the file strictly for streaming
parquet_file = pq.ParquetFile(INPUT_FILE)
print(f"Total rows to process: {parquet_file.metadata.num_rows}")

writer = None
rows_processed = 0

# 2. Iterate in small, fixed-size batches
# use_threads=False reduces memory overhead even further
for batch in parquet_file.iter_batches(batch_size=BATCH_SIZE, use_threads=False):
    
    # Convert pyarrow batch to pandas for processing
    df_chunk = batch.to_pandas()
    
    # Identify numeric columns
    numeric_cols = df_chunk.select_dtypes(include=[np.number]).columns
    
    # --- CLEANUP OPERATIONS (Safe on small chunks) ---
    # 1. Replace INF with NaN
    df_chunk[numeric_cols] = df_chunk[numeric_cols].replace([np.inf, -np.inf], np.nan)
    
    # 2. Fill NaN with 0
    df_chunk[numeric_cols] = df_chunk[numeric_cols].fillna(0)
    
    # 3. Downcast float64 to float32 (Crucial for final file size)
    for col in numeric_cols:
        if df_chunk[col].dtype == 'float64':
            df_chunk[col] = df_chunk[col].astype('float32')

    # Convert back to Arrow Table
    table_chunk = pa.Table.from_pandas(df_chunk)
    
    # Initialize writer with the schema of the first chunk
    if writer is None:
        writer = pq.ParquetWriter(OUTPUT_FILE, table_chunk.schema)
    
    # Write and flush
    writer.write_table(table_chunk)
    
    rows_processed += len(df_chunk)
    print(f"   - Processed {rows_processed} / {parquet_file.metadata.num_rows} rows...")
    
    # FORCE MEMORY CLEAR
    del df_chunk
    del table_chunk
    del batch
    gc.collect() # Force Python to release RAM immediately

if writer:
    writer.close()

print(f"\n✅ SUCCESS: Final dataset saved to '{OUTPUT_FILE}'")
print(f"   - Total Rows: {rows_processed}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib
from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_curve, 
    roc_auc_score, precision_recall_curve, average_precision_score, f1_score
)
from imblearn.ensemble import BalancedRandomForestClassifier
import warnings
warnings.filterwarnings("ignore")

# ==========================
# CONFIGURATION
# ==========================
# Use your final, memory-optimized 300-feature dataset
DATA_FILE = "final_arsenal_300_features_OPTIMIZED.parquet"
OUTPUT_FILE = "final_solar_flare_model_production.joblib"

GROUP_COL = "HARPNUM"
LABEL_COL = "flare"
RANDOM_STATE = 42
TEST_GROUP_FRAC = 0.1  # 10% held out for final testing

# BEST PARAMETERS (From your Optuna Trial 14)
BEST_PARAMS = {
    'n_estimators': 450,
    'max_depth': 24,
    'max_features': 0.116,       # ~11.6% of features per split
    'min_samples_leaf': 8,
    'sampling_strategy': 0.8,    # Ratio of Minority/Majority (Less aggressive than 1.0)
    'n_jobs': -1,
    'random_state': RANDOM_STATE
}

# ==========================
# 1. LOAD DATA
# ==========================
print(f"\n--- Loading dataset: {DATA_FILE} ---")
df = pd.read_parquet(DATA_FILE)

# Ensure label is int
if LABEL_COL not in df.columns and 'classification' in df.columns:
    mapping = {'Non-flare': 0, 'X': 1, 'M': 1, 'C': 0, 'B': 0}
    df[LABEL_COL] = df['classification'].map(mapping).fillna(0).astype(int)

# Select Numeric Features
exclude_cols = {"T_REC", GROUP_COL, "NOAA_AR", "classification", LABEL_COL}
feat_cols = [c for c in df.columns if c not in exclude_cols and pd.api.types.is_numeric_dtype(df[c])]

print(f"Loaded {len(df):,} rows.")
print(f"Features: {len(feat_cols)}")
print(f"Positive Samples: {df[LABEL_COL].sum()}")

# ==========================
# 2. TRAIN/TEST SPLIT
# ==========================
print(f"\n--- Splitting Data (Group-Aware) ---")
gss = GroupShuffleSplit(n_splits=1, test_size=TEST_GROUP_FRAC, random_state=RANDOM_STATE)
train_idx, test_idx = next(gss.split(df, df[LABEL_COL], df[GROUP_COL]))

train_df = df.iloc[train_idx]
test_df = df.iloc[test_idx]

X_train = train_df[feat_cols]
y_train = train_df[LABEL_COL]
X_test = test_df[feat_cols]
y_test = test_df[LABEL_COL]

print(f"Training on {len(X_train):,} samples.")
print(f"Testing on  {len(X_test):,} samples.")

# ==========================
# 3. TRAINING
# ==========================
print(f"\n--- Training Final Balanced Random Forest ---")
print(f"Parameters: {BEST_PARAMS}")

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train
model = BalancedRandomForestClassifier(**BEST_PARAMS)
model.fit(X_train_scaled, y_train)
print("✅ Training Complete.")

# ==========================
# 4. THRESHOLD OPTIMIZATION (The Critical Step)
# ==========================
print(f"\n--- Optimizing Decision Threshold (Maximizing F1-Score) ---")
y_probs = model.predict_proba(X_test_scaled)[:, 1]

# Calculate Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_test, y_probs)

# Calculate F1 Score for every possible threshold
# F1 = 2 * (Precision * Recall) / (Precision + Recall)
f1_scores = 2 * (precision * recall) / (precision + recall)
f1_scores = np.nan_to_num(f1_scores) # Handle division by zero

# Find the best threshold
best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]
best_f1 = f1_scores[best_idx]

print(f"🏆 Optimal Threshold Found: {best_threshold:.4f}")
print(f"🏆 Max F1-Score: {best_f1:.4f}")

# ==========================
# 5. FINAL EVALUATION
# ==========================
print(f"\n--- Final Evaluation at Threshold {best_threshold:.4f} ---")

# Apply the optimal threshold
y_pred_opt = (y_probs >= best_threshold).astype(int)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_opt)
print("Confusion Matrix:\n", cm)

# Classification Report
print("\nClassification Report:\n", classification_report(y_test, y_pred_opt, digits=4))

# Metrics
tn, fp, fn, tp = cm.ravel()
tss = (tp / (tp + fn)) - (fp / (fp + tn))
roc = roc_auc_score(y_test, y_probs)
pr_auc = average_precision_score(y_test, y_probs)

print(f"Final TSS:     {tss:.4f}")
print(f"Final ROC AUC: {roc:.4f}")
print(f"Final PR AUC:  {pr_auc:.4f}")

# ==========================
# 6. PLOTTING
# ==========================
print("\n--- Generating Plots ---")

# A. Feature Importance
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]
top_n = 20
plt.figure(figsize=(12, 6))
plt.title(f"Top {top_n} Features (Final Model)")
plt.bar(range(top_n), importances[indices][:top_n], align="center")
plt.xticks(range(top_n), np.array(feat_cols)[indices][:top_n], rotation=45, ha="right")
plt.tight_layout()
plt.show()

# B. Threshold Tuning Curve
plt.figure(figsize=(10, 6))
plt.plot(thresholds, f1_scores[:-1], label="F1 Score")
plt.plot(thresholds, precision[:-1], label="Precision", linestyle='--')
plt.plot(thresholds, recall[:-1], label="Recall", linestyle='--')
plt.axvline(best_threshold, color='red', linestyle=':', label=f"Best Threshold ({best_threshold:.2f})")
plt.xlabel("Probability Threshold")
plt.ylabel("Score")
plt.title("Threshold Optimization Curve")
plt.legend()
plt.grid(True)
plt.show()

# ==========================
# 7. SAVE EVERYTHING
# ==========================
joblib.dump({
    "model": model,
    "scaler": scaler,
    "features": feat_cols,
    "threshold": best_threshold,
    "params": BEST_PARAMS
}, OUTPUT_FILE)
print(f"\n✅ Model saved to {OUTPUT_FILE}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import confusion_matrix, classification_report, roc_curve

# =========================
# CONFIGURATION
# =========================
MODEL_FILE = "final_solar_flare_model_production.joblib"
DATA_FILE = "final_arsenal_300_features_OPTIMIZED.parquet"
LABEL_COL = "flare"
GROUP_COL = "HARPNUM"
TEST_GROUP_FRAC = 0.1
RANDOM_STATE = 42

# CHANGE THIS to pick your operational point (0.25 is usually the sweet spot)
SELECTED_THRESHOLD = 0.25 

# =========================
# 1. LOAD MODEL & DATA
# =========================
print(f"--- Loading Model and Data ---")
bundle = joblib.load(MODEL_FILE)
model = bundle['model']
scaler = bundle['scaler']
feat_cols = bundle['features']

df = pd.read_parquet(DATA_FILE)

# Ensure Label
if LABEL_COL not in df.columns and 'classification' in df.columns:
    mapping = {'Non-flare': 0, 'X': 1, 'M': 1, 'C': 0, 'B': 0}
    df[LABEL_COL] = df['classification'].map(mapping).fillna(0).astype(int)

# =========================
# 2. RECREATE TEST SET
# =========================
gss = GroupShuffleSplit(n_splits=1, test_size=TEST_GROUP_FRAC, random_state=RANDOM_STATE)
train_idx, test_idx = next(gss.split(df, df[LABEL_COL], df[GROUP_COL]))
test_df = df.iloc[test_idx]

X_test = test_df[feat_cols]
y_test = test_df[LABEL_COL]

# Scale
X_test_scaled = scaler.transform(X_test)
y_probs = model.predict_proba(X_test_scaled)[:, 1]

# =========================
# 3. PRINT MENU OF OPTIONS
# =========================
print("\n" + "="*85)
print(f"{'THRESHOLD':<10} | {'RECALL':<10} | {'PRECISION':<10} | {'TSS':<10} | {'FALSE ALARMS':<15} | {'MISSED FLARES':<15}")
print("="*85)

thresholds = [0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.50]

for thresh in thresholds:
    y_pred_t = (y_probs >= thresh).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_t).ravel()
    
    recall = tp / (tp + fn)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    tss = recall - (fp / (fp + tn))
    
    marker = "   <-- SELECTED" if thresh == SELECTED_THRESHOLD else ""
    print(f"{thresh:<10.2f} | {recall:<10.4f} | {precision:<10.4f} | {tss:<10.4f} | {fp:<15} | {fn:<15}{marker}")

print("="*85)

# =========================
# 4. FINAL EVALUATION (AT SELECTED THRESHOLD)
# =========================
print(f"\n--- Detailed Report at Threshold {SELECTED_THRESHOLD} ---")
y_pred = (y_probs >= SELECTED_THRESHOLD).astype(int)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=4))

# Visualize
fpr, tpr, roc_thresholds = roc_curve(y_test, y_probs)
tss_scores = tpr - fpr

plt.figure(figsize=(10, 6))
plt.plot(roc_thresholds, tss_scores, label="TSS Curve", color='blue')
plt.axvline(SELECTED_THRESHOLD, color='green', linestyle='--', label=f"Selected ({SELECTED_THRESHOLD})")
plt.xlabel("Threshold")
plt.ylabel("TSS Score")
plt.title("Selecting the Operational Threshold")
plt.legend()
plt.grid(True)
plt.show()
```
